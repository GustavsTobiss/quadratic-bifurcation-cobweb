```python
import numpy as np
import matplotlib.pyplot as plt
import math as math

#paramater to change in the parabola; the property of fixed points changes drastically with c
c=0.25
#starting seed
x_0=0
#Number that determines the refinement of the interval [-2,0.25]
N=1000

#arrays for various data
x_points=[x_0] #stores the iterates of x_0
fixed_points=[] #stores the fixed points
fixed_points_temp=[] #holds temporary values
x=[] #stores final points to plot along x axis (c values)
y=[] #stores final points to plot along y axis (x values whivh are n-cycles)

#iterates the x**2. This can be changed to make diagrams for any function.
def fa(a):
    b=int(100000*((a**2)+c))/100000
    return(b)

#plotty things
f = plt.figure()
f.set_figwidth(25)
f.set_figheight(20)
plt.xlabel ('c',fontsize=35)
plt.ylabel ('x',fontsize=35, rotation=90)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.plot([-20,20],[0,0],color="lightgray")
plt.plot([0,0],[-20,20],color="lightgray")
plt.axis([-2,1/4,-2,2])


for j in range(0,N+1):
    c=0.25-(j*2.25)/N #increments c
    for i in range(0,500): #iterates the seed 5000 times and stores it in a list
        x_points.append(fa(x_points[i]))
    x_points=np.array(x_points) #convert to np array for easy array manipulation
    x_points=np.unique(x_points) #removes any duplicates (and orders them by size but that doesn't matter)
    x_points=x_points.tolist() #converts back to python list for faster computations
    fixed_points_temp=[] #wipes the temp list
    for k in range(0,len(x_points)): #this whole loop is for sifting through iterates to find cycles/fixed points
        temp=x_points[k] #takes an iterate to test
        for m in range(0,N): #iterates the test a lot of times
            temp=fa(temp) 
            if temp == x_points[k]: #at each iterate tests if the iterate is a loop
                fixed_points_temp.append(temp)
        fixed_points_temp=np.array(fixed_points_temp) #more list fiddling
        fixed_points_temp=np.unique(fixed_points_temp)
        fixed_points_temp=fixed_points_temp.tolist()
    for k in range(0,len(fixed_points_temp)): #stores each seed wich admits an n-cycle with the associated c value
        fixed_points.append([c,fixed_points_temp[k]])


#puts the data into arrays which are easier to graph
for i in range(0,len(fixed_points)):
    x.append(fixed_points[i][0])
    y.append(fixed_points[i][1])

plt.scatter(x,y,s=0.2,color="blue")
```
