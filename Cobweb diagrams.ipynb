```python
import numpy as np
import matplotlib.pyplot as plt
import math as math
#used to set the fineness of the x axis; make this a power of 10
refinement=100000
#cut off for x values; i.e. abs(x)<maximum
maximum=12
#creates the x axis array
x=np.linspace(-maximum,maximum,maximum*(refinement)+1)
#lists to store values for the cobweb diagram
x_points=[]
y_points=[]
#lists to store the fixed point values
fixed_points_x=[]
fixed_points_y=[]
#paramater to change in the parabola; the property of fixed points changes drastically with c
c=-2.1
#starting seed
x_0=1

#iterates the x**2. This can be changed to make diagrams for any function.
def fa(a):
    return((a*a)+c)
#makes a list which stores the values of the function 
x2=fa(x)

#plotting garbage
f = plt.figure()
plt.title('x_0 = '+str(x_0)+', c = '+str(c),fontsize=50)
f.set_figwidth(20)
f.set_figheight(20)
plt.plot([-maximum,maximum],[0,0],color="lightgray")
plt.plot([0,0],[-maximum,maximum],color="lightgray")
plt.plot([-maximum,maximum],[-maximum,maximum],color="gray")
a=3 #change this to change the zoom on the graph
plt.axis([-a,a,-a,a])
# finds the (approximate) fixed points 
for i in range(0,len(x-1)):
    if abs(x[i]-x2[i])<10**(-4):
        fixed_points_x.append(x[i])
        fixed_points_y.append(x2[i])

#plots (seed, iterated seed)
plt.scatter(x_points,y_points,s=50)
#plots parabola
plt.plot(x,x2)
#initializes data
x_points.append(x_0)
y_points.append(fa(x_0))

#does the logic for cobwebbing
for i in range(0,100): 
    x_points.append(y_points[i])
    y_points.append(fa(x_points[i]))
    if y_points[i]>10**5: #failsafe in case orbits diverge
        break
#more plotting garbage
plt.scatter(fixed_points_x,fixed_points_y,color="red",s=100)
plt.plot(x_points,y_points)
plt.show()
```
